```python
#Extract the file from the downloaded zip file
zipfile_name = 'london-bike-sharing-dataset.zip'
with zipfile.ZipFile(zipfile_name, 'r') as file:
    file.extractall()
```

```python
#Read the file as pd frame
bikes = pd.read_csv('london_merged.csv')
```

```python
#Rename the columns to the specificed column names
new_col_dict = {
    'timestamp': 'time',
    'cnt': "count",
    't1': 'temp_real_C',
    't2': 'temp_feel_like_C',
    'hum': 'humidity_percent',
    'wind_speed': 'wind_speed_kph',
    'weather_code': 'weather',
    'is_holiday':'is_holiday',
    'is_weekend':'is_weekend',
    'season':'season'
}

bikes.rename(new_col_dict, axis=1, inplace=True)
```

```python
#Reformat the humidity to precentages
bikes.humidity_percent = bikes.humidity_percent/100
```

```python
#Create a season dictionary to map the integers to the actual written values
season_dict = {
    '0.0': 'spring',
    '1.0': 'summer',
    '2.0': 'autumn',
    '3.0': 'winter'
}

#Create a weather dictionary to map the integers to the actual written values
weather_dict = {
    '1.0':'Clear',
    '2.0':'Scattered clouds',
    '3.0':'Broken clouds',
    '4.0':'Cloudy',
    '7.0':'Rain',
    '10.0':'Rain with thunderstorm',
    '26.0':'Snowfall'
}

#Change the type and mapping of the season column
bikes.season = bikes.season.astype('str')
bikes.season = bikes.season.map(season_dict)

#Change the type and mapping of the season column
bikes.weather = bikes.weather.astype('str')
bikes.weather = bikes.weather.map(weather_dict)
```
